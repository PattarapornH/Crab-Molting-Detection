```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from google.colab import drive
drive.mount('/content/gdrive')
```

```
path_train = 'gdrive/My Drive/Year4/Project/dataset/'
path_molting_train = []
path_normal_train = []
for path in os.listdir(path_train + 'class_0'):
  path_molting_train.append(os.path.join(path_train+'class_0', path))
for path in os.listdir(path_train + 'class_1'):
  path_normal_train.append(os.path.join(path_train+'class_1', path))
```

```
import cv2 as cv

img_molt = cv.imread(path_molting_train[0])
img_molt = cv.resize(img_molt,(500,500))

img_normal = cv.imread(path_normal_train[0])
img_normal = cv.resize(img_normal,(500,500))

plt.subplot(121)
plt.title('molting')
plt.imshow(img_molt)

plt.subplot(122)
plt.title('normal')
plt.imshow(img_normal)
```

```
edge_molt = cv.Canny(img_molt,100,200)
edge_normal = cv.Canny(img_normal,100,200)

plt.subplot(121)
plt.title('molting')
plt.imshow(edge_molt)

plt.subplot(122)
plt.title('normal')
plt.imshow(edge_normal)
```
